```python
import pandas as pd
```

```python
diamonds = pd.read_csv('./train.csv', index_col=0)
```

```python
diamonds.shape
```

```python
diamonds.head()
```

```python
diamonds.cut.unique()
```

### Given that this characteristic can be replaced by numbers let's do that

```python
diamonds['cut'] = diamonds['cut'].replace(['Fair','Good', 'Very Good', 'Ideal', 'Premium'],[1, 2, 3, 4, 5])
```

```python
diamonds.head()
```

### Also, I have looked up that for colors, the lower the grade scale the better, let's put that into numeric values also

```python
diamonds.color.unique()
```

```python
diamonds['color'] = diamonds['color'].replace(['J','I', 'H', 'G', 'F', 'E', 'D'],[1, 2, 3, 4, 5, 6, 7])
```

```python
diamonds.head()
```

```python
diamonds.drop(['clarity'], axis=1, inplace=True)
```

### Model

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
gs = GridSearchCV(
    estimator=DecisionTreeRegressor(),
    param_grid={
        "max_depth": [5, 6],
        "min_samples_split": [40, 120, 400, 1000],
        "max_features": [7, 9]
    },
    cv=5,
    verbose=3,
    scoring="neg_mean_squared_error",
    return_train_score=True
)
```

```python
X = diamonds.drop('price', axis=1)
```

```python
y = diamonds.price
```

```python
%%time
gs.fit(X, y)
```

```python
grid_search_results = pd.DataFrame(gs.cv_results_)
grid_search_results = grid_search_results[['param_max_depth', 'param_max_features', 'param_min_samples_split',
       'mean_test_score', 'mean_train_score']]
```

```python
grid_search_results.sort_values("mean_test_score", ascending=False).head(10)
```

```python
best_tree = gs.best_estimator_
```

```python
best_tree
```

### Predict test

```python
test = pd.read_csv("./test.csv", index_col=0)
```

```python
test.head()
```

```python
test['cut'] = test['cut'].replace(['Fair','Good', 'Very Good', 'Ideal', 'Premium'],[1, 2, 3, 4, 5])
```

```python
test['color'] = test['color'].replace(['J','I', 'H', 'G', 'F', 'E', 'D'],[1, 2, 3, 4, 5, 6, 7])
```

```python
test.drop(['clarity'], axis=1, inplace=True)
```

```python
y_pred = best_tree.predict(test)
```

```python
test["price"] = y_pred
```
